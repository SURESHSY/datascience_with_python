```python
#building dashboards
sns.set_style("dark",{"axes.facecolor":"black"})
f,axes = plt.subplots(2,2,figsize=(15,15))
k1 = sns.kdeplot(movieData.Budget,movieData.audienceRating,ax=axes[0,0],shade=True,shade_lowest=False,cmap="rainbow")

k2 = sns.kdeplot(movieData.Budget,movieData.criticRating,ax=axes[0,1],shade=True,shade_lowest=False,cmap="inferno")

v1 =sns.violinplot(data=movieData,x="Genre",y="criticRating",ax=axes[1,0])

kde1 = sns.kdeplot(movieData.criticRating,movieData.audienceRating,shade=True,shade_lowest=False,cmap="Blues_r",ax=axes[1,1])

kde2 = sns.kdeplot(movieData.criticRating,movieData.audienceRating,cmap="gist_gray_r",ax=axes[1,1])
k1.set(xlim=(-20,160))
k2.set(xlim=(-20,160))
#same as building subplots
```
